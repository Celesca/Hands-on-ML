```python
df2 = pd.read_csv('Jan-17-2024_Mar-31-2024.csv')
```

```python
new_df = df2[['ชื่อเพจ', 'ชื่อ', 'เวลาที่เผยแพร่' , 'อิมเพรสชั่น',  'ระยะเวลา (วินาที)', 'การเข้าถึง', 'การรับชมวิดีโอนาน 3 วินาที', 'การรับชมวิดีโอนาน 1 นาที', 'จำนวนวินาทีที่รับชมโดยเฉลี่ย', 'ความรู้สึก ความคิดเห็น และการแชร์']]
new_df
```

```python
# sort the new_df with the column 'อิมเพรสชั่น' in descending order
new_df = new_df.sort_values(by='อิมเพรสชั่น', ascending=False)
new_df
```

```python
# Sum the เอม เพรสชั่น และ จำนวนวินาทีที่รับชมโดยเฉลี่ย by 'ชื่อ' with name of each games such as "OW2", "Valorant", "Goose Goose Duck", "Blazing 8s", "LoL" 
grouped_df = new_df[new_df['ชื่อ'].str.contains('OW2|Valorant|Goose Goose Duck|Blazing 8s|LoL|TFT', na=False)].groupby('ชื่อ').agg({
    'อิมเพรสชั่น': 'sum',
    'จำนวนวินาทีที่รับชมโดยเฉลี่ย': 'sum',
    'ความรู้สึก ความคิดเห็น และการแชร์': 'sum'
}).reset_index()
grouped_df
```

```python
import pandas as pd

# Define the list of games
games = ['OW2', 'Valorant', 'Goose Goose Duck', 'Blazing 8s', 'LoL', 'TFT', 'RoV']

# Replace any 'ชื่อ' containing the game names with the game name itself
for game in games:
    new_df['ชื่อ'] = new_df['ชื่อ'].str.replace(rf'.*{game}.*', game, regex=True)

# Group by the updated 'ชื่อ' and aggregate the required columns
grouped_df = new_df[new_df['ชื่อ'].str.contains('|'.join(games), na=False)].groupby('ชื่อ').agg({
    'อิมเพรสชั่น': 'sum',
    'จำนวนวินาทีที่รับชมโดยเฉลี่ย': 'sum',
    'ความรู้สึก ความคิดเห็น และการแชร์': 'sum'
}).reset_index()

grouped_df
```

```python
grouped_df = grouped_df.sort_values(by='จำนวนวินาทีที่รับชมโดยเฉลี่ย', ascending=False)
grouped_df
```
